```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_all = pd.read_excel("sp100new.xlsx",sheet_name = "Data", index_col=0)
weight_all = pd.read_excel("sp100new.xlsx",sheet_name = "Portfolio", index_col=0)
```

```python
data_all.head()
```

```python
weight_all.head()
```

```python
data_return = data_all.apply(lambda x: np.log(x/x.shift(1)).dropna())
data_return.head()
```

```python
pf_return = (data_return*weight_all).sum(axis=1) / 1000000  # because of leveraged position

pf_return = pf_return.iloc[1:]
pf_return
```

## A. Basic: Return, risk and correlation 

```python
pf_week1 = pf_return.iloc[:5]
week_ret1 = np.cumprod(1+pf_week1).iloc[-1]-1
print("First week expected return (annualized): ", round(week_ret1,4))
pf_week2 = pf_return.iloc[5:10]
week_ret2 = np.cumprod(1+pf_week2).iloc[-1]-1
print("Second week expected return (annualized): ", round(week_ret2,4))
pf_week3 = pf_return.iloc[10:]
week_ret3 = np.cumprod(1+pf_week3).iloc[-1]-1
print("Third week expected return (annualized): ", round(week_ret3,4))
```

```python
annu_return = np.cumprod(1+pf_return).iloc[-1]-1
print("Annualized expected return: ", round(annu_return,4))
```

Return: expected return

The performance of this strategy and the efficacy of signals are good as total annualized expected return are positive, and the momentum signals tend to make positive predictions under the backtest period. The timing of alpha is also acceptable because although return of first week is relatively low, starting from second week, it can generate larger postive returns. 

Meanwhile, since this strategy is not of high-frequency, the transaction cost is relatively low. But because of involving short selling stocks, there are financing costs that need to be considered, making the strategy less attractable. 

```python
def cal_IR(arr):
    
    """Information Ratio"""
    
    annu_return = np.cumprod(1+arr).iloc[-1]-1
    annu_vol = np.std(arr)*np.sqrt(250)
    return annu_return / annu_vol

IR = cal_IR(pf_return)
print("Information ratio (IR): ", round(IR,4))
annu_vol = np.std(pf_return)*np.sqrt(250)
print("Annualized volatility: ", round(annu_vol,4))
skewness = pf_return.skew()
print("Daily Skewness: ", round(skewness, 4))
```

Risk: IR = 0.3759

The risk adjusted performance is assessed by information ratio, which is annualized return divided by annualized volatility of the portfolio. The original selection of stocks are discretional (by choosing top 100 market cap stocks of s&p 500 constituents), hence we foresee IR to be low. The annualized volatility is 21.89% which is slightly higher than 20% and hence showing that the portfolio is volatile and correlation and volatility risk have not been fully hedged. 

In terms of higher moment, the daily skewness is calculated as 0.7621 which is larger than 0.5, meaning the portfolio is slightly positive skewed and hence we are likely to have mean of daily return larger than median for portfolio return which can generate more favorable returns than a symmetric portfolio.

```python
spy = pd.read_excel("sp100new.xlsx",sheet_name = "SPY", index_col=0)["Adj Close"]
spy_return = np.log(spy/spy.shift(1)).dropna()

corr_spy = np.corrcoef(pf_return, spy_return)
print("Correlation with SPY: ", round(corr_spy[1][0],4))
```

## B: Loss statistics

```python
sorted_pf_ret = pf_return.sort_values()
n = len(sorted_pf_ret)
VaR = sorted_pf_ret.iloc[int(np.ceil(1-0.75*n))]
dollarVaR = 2000000 * VaR
print("75% daily VaR is {:.4f}.".format(VaR))
print("75% dollar VaR is {:.4f}".format(dollarVaR))
ETL = sorted_pf_ret.iloc[:int(np.ceil(1-0.75*n))].mean()
dollarETL = 2000000 * ETL
print("75% ETL is {:.4f}.".format(ETL))
print("75% dollar ETL is {:.4f}".format(dollarETL))
```

75\% VaR of the portfolio is calculated using a historical method by ranking all 14 observations of portfolio returns from the largest loss to the largest profit, and the 25\% quantile of the largest loss is the 75% percent VaR and dollar VaR is obtained by multiplying percent VaR and gross value of the portfolio (\$2000000)

75\% ETL of the portfolio is calculated using a historical method by calculating the average loss of all situations worse than 75\% VaR, and dollar ETL is obtained by multiplying percent ETL and gross value of the portfolio (\$2000000)

```python
def MaxDrawdown(arr):
    '''max drawdown'''
    i = np.argmax((np.maximum.accumulate(arr) - arr) / np.maximum.accumulate(arr))#结束位置
    if i == 0:
        return 0
    j = np.argmax(arr[:i])  # 开始位置
    maxdrawdown = (arr[j] - arr[i]) / (arr[j]) * 100  # percent
    return maxdrawdown, j, i

pf_value = np.cumprod(1+pf_return)
maxdd, start, end = MaxDrawdown(pf_value)
print("Max peak-to-trough drawdown is {:.2f}%, starting from {} and end on {}.".format(maxdd,start.date(),end.date()))
```
